```python
import geopy.distance
import math
def get_distance_point(lat, lon, distance, direction):
    """
    根据经纬度，距离，方向获得一个地点
    :param lat: 纬度
    :param lon: 经度
    :param distance: 距离（千米）
    :param direction: 方向（北：0，东：90，南：180，西：360）
    :return:
    """
    start = geopy.Point(lat, lon)
    d = geopy.distance.VincentyDistance(kilometers=distance)
    return d.destination(point=start, bearing=direction)
def get_squareofpoint(lat,lon,distance=500):
    sq1 = get_distance_point(lat,lon,distance/1000*math.sqrt(2),45)
    sq2 = get_distance_point(lat,lon,distance/1000*math.sqrt(2),135)
    sq3 = get_distance_point(lat,lon,distance/1000*math.sqrt(2),225)
    sq4 = get_distance_point(lat,lon,distance/1000*math.sqrt(2),315)
    return [sq1,sq2,sq3,sq4]
from urllib.parse import quote
from urllib import request
import json
 
# TODO
amap_web_key = '323767bdd936d9ac903cb0a75210aff8'  # 高德地图官网申请的Web API KEY
 

# POI分类集合
types = {"010000":"汽车服务",
         "020000":"汽车销售",
         "030000":"汽车维修",
         "040000":"摩托车服务",
         "050000":"餐饮服务",
         "060000":"购物服务",
         "070000":"生活服务",
         "080000":"体育休闲服务",
         "090000":"医疗保健服务",
         "100000":"住宿服务",
         "110000":"风景名胜",
         "120000":"商务住宅",
         "130000":"政府机构及社会团体",
         "140000":"科教文化服务",
         "150000":"交通设施服务",
         "160000":"金融保险服务",
         "170000":"公司企业",
         "180000":"道路附属设施",
         "190000":"地名地址信息",
         "200000":"公共设施",
         "220000":"事件活动",
         "970000":"室内设施",
         "990000":"通行设施"      
        }
types_houses={
    "120100":"产业园区",
    "120200":"楼宇相关",
    "120300":"住宅区",

}
poi_search_url = "http://restapi.amap.com/v3/place/around"
offset=25  # 分页请求数据时的单页大小
# 单页获取pois
'''
?key=您的key&location=116.473168,39.993015&keywords=&types=050000&radius=1000&offset=20&page=1&extensions=all
'''
def getpoi_page(lat, lon, types, radius):
    req_url = poi_search_url + "?key=" + amap_web_key + '&location=' + lat + "," + lon + '&keywords=&types=' + types + '&radius=' + radius + '&offset=20&page=1&extensions=all'
    data = ''
    with request.urlopen(req_url) as f:
        data = f.read()
        data = data.decode('utf-8')
    return data
def getpoi_count(lat, lon, types, radius):
    result = getpoi_page(lat, lon, types, radius)
    result = json.loads(result)
    return result
```
